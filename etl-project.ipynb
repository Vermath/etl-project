```python
import pandas as pd
import sqlalchemy
from sqlalchemy import create_engine
```

```python
#importing files
csv_file1 = "accidental_shootings.csv"
accidental_shootings_df = pd.read_csv(csv_file1)
accidental_shootings_df.head()

#cleaning
accidental_shootings_df = accidental_shootings_df[['Incident ID', 'Incident Date', 'State', 'City Or County', '# Killed', '# Injured']].copy()

clean_accidental_shootings_df = accidental_shootings_df.rename(columns={"Incident ID":"Incident_ID", "Incident Date":"Incident_Date", "State":"State", "City Or County":"CityOrCounty", "# Killed":"Num_Killed", "# Injured":"Num_Injured"})
clean_accidental_shootings_df.head()
```

```python
#importing files
csv_file1 = "mass_shootings.csv"
mass_shootings_df = pd.read_csv(csv_file1)
mass_shootings_df.head()

#cleaning
mass_shootings_df = mass_shootings_df[['Incident ID', 'Incident Date', 'State', 'City Or County', '# Killed', '# Injured']].copy()

clean_mass_shootings_df = mass_shootings_df.rename(columns={"Incident ID":"Incident_ID", "Incident Date":"Incident_Date", "State":"State", "City Or County":"CityOrCounty", "# Killed":"Num_Killed", "# Injured":"Num_Injured"})
clean_mass_shootings_df.head()
```

```python
#sql junk, creating engine, checking table names
engine = create_engine('postgres://postgres:Cable9630!@localhost:5432/etl_challenge')
engine.table_names()
```

```python
#loading CSV to database
clean_mass_shootings_df.to_sql(name='mass_shootings', con=engine, 
                              if_exists='append', index=False)

clean_accidental_shootings_df.to_sql(name='accidental_shootings', con=engine, 
                              if_exists='append', index=False)
```

```python
#checking to make sure everything is working
pd.read_sql_query('select * from mass_shootings', con=engine).head()
```

```python
#checking to make sure everything is working
pd.read_sql_query('select * from accidental_shootings', con=engine).head()
```
